```python
import os
import numpy as np 
import pandas as pd 
from subprocess import check_output
from time import localtime, strftime
#print(check_output(["ls", "../submit"]).decode("utf8"))
```

```python
sub_path = "../submit/Best Results"
all_files = os.listdir(sub_path)

# Read and concatenate submissions
outs = [pd.read_csv(os.path.join(sub_path, f), index_col=0) for f in all_files]
concat_sub = pd.concat(outs, axis=1)
cols = list(map(lambda x: "is_iceberg_" + str(x), range(len(concat_sub.columns))))
concat_sub.columns = cols
concat_sub.reset_index(inplace=True)
concat_sub.head()
```

```python
#Check Correlation
concat_sub.corr()
```

```python
# get the data fields ready for stacking
n = len(concat_sub.columns)
concat_sub['is_iceberg_max'] = concat_sub.iloc[:, 1:n].max(axis=1)
concat_sub['is_iceberg_min'] = concat_sub.iloc[:, 1:n].min(axis=1)
concat_sub['is_iceberg_mean'] = concat_sub.iloc[:, 1:n].mean(axis=1)
concat_sub['is_iceberg_median'] = concat_sub.iloc[:, 1:n].median(axis=1)
concat_sub.head()
```

```python
# set up cutoff threshold for lower and upper bounds, easy to twist 
cutoff_lo = 0.7
cutoff_hi = 0.3
```

## MinMax + BestBase Stacking

```python
# load the model with best base performance
sub_base = pd.read_csv('../submit/Best Results/stack_minmax_bestbase.csv')
# name string
name_str = strftime("%Y%m%d%H%M", localtime())
concat_sub['is_iceberg_base'] = sub_base['is_iceberg']
concat_sub['is_iceberg'] = np.where(np.sum(concat_sub.iloc[:,1:n] > cutoff_lo, axis=1)>(0.7*(n-1)), 
                                    concat_sub['is_iceberg_max'], 
                                    np.where(np.sum(concat_sub.iloc[:,1:n] < cutoff_hi, axis=1)>=(0.7*(n-1)),
                                             concat_sub['is_iceberg_min'], 
                                             concat_sub['is_iceberg_base']))
concat_sub[['id', 'is_iceberg']].to_csv('stacking'+name_str+'.csv', 
                                        index=False, float_format='%.6f')
```
